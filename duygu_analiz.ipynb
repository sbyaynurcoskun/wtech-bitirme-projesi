```python
def islem(review):
    review=BeautifulSoup(review).get_text()  #tagları sildim
    review=re.sub("[^a-zA-Z]", ' ' , review) #harf dışı sildim
    review.lower()   #büyük küçük harf
    review=review.split()  #kelimelere böldüm kök
    swords=set(stopwords.words("english"))
    review=[w for w in  review if w in swords]
    return(" ".join(review))
```

```python
train_x_tum=[]  #liste
for i in range(len(data["review"])):
    if(i+1)%1000==0:  #1k nın katlarında haber vercek
        print("işlem adımı" , i+1)
    train_x_tum.append(islem(data["review"][i]))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=train_x_tum  #verileri aldık
y=np.array(data["sentiment"])  #veri içindeki 0-1ler(o'lu o'suz)
```

```python
data
```

```python
train_x,test_x,train_y,test_y=  train_test_split(x,y,test_size=0.15,random_state=33)
```

```python
from sklearn.feature_extraction.text import CountVectorizer #kelimeleri sayısal metine dönderdim (vektörleştirme)
```

```python
vectorizer=CountVectorizer(max_features=5000)
```

```python
train_x1=vectorizer.fit_transform(train_x) #x'i transforme edip x1'in içine attım
```

```python
test_x1=vectorizer.transform(test_x)
```

```python
train_x1  #matris boyut
```

```python
train_x1=train_x1.toarray() #matrisi diziye çevirdim
```

```python
train_y1=train_y  #y1 tanımladım
```

```python
train_x1.shape,train_y1.shape  #eşitmi baktım
```

```python
test_x1=test_x1.toarray()
```
